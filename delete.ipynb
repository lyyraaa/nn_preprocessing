```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
unet_mf1s = [.8744,.882919,.882083,.84103,.882733,.57813]
unet_vars = [.041295,.027193,.024837,0.040272,.023042,.163211]
unet_names = ["unet_3","unet_4","unet_5","attention_unet","feature_enhance","unext"]

model_mf1s = [0.745023,0.73319,.772422,.837328,.848713,.900975,.882919]
model_vars = [.031608,.034415,.036484,.031825,.036201,.033395,.027103]
model_names = ["rf","svm","patch_3px","patch_25px","patch_101px","patch_3+101px","unet"]
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,6))
cmap = plt.get_cmap('cividis')
lines = {'linestyle': 'None'}
plt.rc('lines', **lines)
ax[0].bar(np.arange(0, len(model_mf1s)), model_mf1s,color = [cmap(x / len(model_names))[:-1] for x in range(len(model_names))] )
ax[0].errorbar(np.arange(0, len(model_vars)), model_mf1s, yerr=model_vars, fmt='k')
ax[0].set_xticks(np.arange(0,len(model_mf1s)),labels=model_names,rotation=10,)
ax[0].set_ylim(ymin=0.6,ymax=1.); ax[0].set_ylabel("Macro F1 Score (10-trial avg)")
ax[0].set_title("10-trial performance avg. for all models")

ax[1].bar(np.arange(0, len(unet_mf1s)), unet_mf1s,color = [cmap(x / len(unet_names))[:-1] for x in range(len(unet_names))] )
ax[1].errorbar(np.arange(0, len(unet_vars)), unet_mf1s, yerr=unet_vars, fmt='k')
ax[1].set_xticks(np.arange(0,len(unet_mf1s)),labels=unet_names,rotation=10,)
ax[1].set_ylim(ymin=0.6,ymax=1.); ax[1].set_ylabel("Macro F1 Score (10-trial avg)")
ax[1].set_title("10-trial performance avg. for UNet models")

fig.tight_layout()
```
